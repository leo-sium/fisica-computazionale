Qui faccio un esempio di plot di un grafico del potente sium

```python
import numpy as np

data = np.loadtxt("computazionale/1-esercizio/datif1.dat")

x = data[:, 0]
y = data[:, 1]
```

```python
import matplotlib.pyplot as plt

plt.scatter(x, y, s = 5, label='dati')
plt.xlabel('asse X')
plt.ylabel('asse Y')
plt.xscale("log")
plt.legend()
plt.show()
```
